```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Problem A
hx1 = 'C:\cs612\CS612-Summer-2020\IN\HunarNadha\Programming_exercise2\Heights.csv' 
df = pd.read_csv(hx1)
plt.scatter(df['Husband'], df['Wife'])
plt.title("Scatter plot of Heights")
plt.xlabel('Husband')
plt.ylabel('Wife')
sns.regplot(df['Husband'], df['Wife'])
plt.show()
```

```python
#Problem B
x =np.arange(-5, 4.9, 0.1)
y1 = -(1/3) * x + 1.5
plt.plot(x, y1, color = 'green', ls = 'dotted', lw =5)
plt.legend(["Graph1"], loc = "upper center")
```

```python
y2 = 0.02 * x * (x-1) * (x+1) + 1
plt.plot(x, y2, color = 'yellow', ls = 'dashed', lw=8)
plt.legend(["Graph2"], loc = "upper center")
```

```python
y3 = 2 * np.exp(pow(-x, 2))
plt.plot(x, y3, color = 'red', ls = 'solid', lw=6)
plt.legend(["Graph3"], loc = "upper center")
```

```python
plt.plot(x,y1, label = "Graph1",color='green', lw=11)
plt.plot(x,y2, label = "Graph2",color='yellow', lw=7)
plt.plot(x,y3, label = "Graph3",color='red' , lw=4)
plt.legend(loc = "upper center")
plt.show()
```
