```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-np.pi,np.pi,256,endpoint=True)
c, s = np.cos(x),np.sin(x)
plt.plot(x,c,'-c')
plt.plot(x,s,'--m')
```

```python
plt.plot(x,c,color="cyan",linestyle="-")
plt.plot(x,s,color="magenta",linestyle="--")
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

y = [1,4,9,16,25,36,49,64]
x1 = [1,16,30,42,55,68,77,88]
x2 = [1,6,12,18,28,40,52,65]

l1 = ax.plot(x1,y,'bs-')
l2 = ax.plot(x2,y,'ro--')
ax.legend(labels=('AMZ','MSFT'),loc='lower right')
ax.set_title("Stock Market")
ax.set_xlabel('Time')
ax.set_ylabel('USD')
plt.show()
```

```python
x = np.arange(0,np.pi*2,0.05)
y = np.sin(x)
z = np.cos(x)

fig2 = plt.figure()

axes1 = fig2.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig2.add_axes([0.55,0.55,0.3,0.3])

axes1.plot(x,y,'b')
axes2.plot(x,z,'r')
axes1.set_title('sine')
axes2.set_title('cosine')
plt.show()
```

```python
def f(x,y):return (1-x/2+x**5+y**3)*np.exp(-x**2-y**2)

n = 256
x = np.linspace(-3,3,n)
y = np.linspace(-3,3,n)
X,Y = np.meshgrid(x,y)

plt.contourf(X,Y,f(X,Y),8,alpha=.75,cmap='jet')
C = plt.contour(X,Y,f(X,Y),8,colors='black',linewidth=.5)
plt.show()
```

```python
n = 512
x = np.random.normal(0,1,n)
y = np.random.normal(0,1,n)

plt.scatter(x,y,color='c')
plt.show()
```

```python
n = 12
x = np.arange(n)
y1 = (1-x/float(n)) * np.random.uniform(0.5,1.0,n)
y2 = (1-x/float(n)) * np.random.uniform(0.5,1.0,n)

plt.bar(x,+y1,facecolor='#9999ff',edgecolor=['green','red'])
plt.bar(x,-y2,facecolor='#ff9999',edgecolor='white')

for x,y in zip(x,y1):
    plt.text(x+0.4,y+0.05,f"{y:.2f}",ha='center',va='center')

plt.ylim(-1.25,+1.25)
plt.show()

    
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = Axes3D(fig)

x = np.arange(-4,4,0.25)
y = np.arange(-4,4,0.25)

x, y = np.meshgrid(x,y)
r = np.sqrt(x**2 + y**2)
z = np.sin(r)

ax.plot_surface(x,y,z,rstride=1,cstride=1,cmap='winter')

plt.show()
```
