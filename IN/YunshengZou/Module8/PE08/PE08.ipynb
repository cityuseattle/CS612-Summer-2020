```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Heights.csv')
data
```

```python
#Problem A
ax = plt.subplot()
x = data['Husband']
y = data['Wife']
ax.scatter(x,y)
ax.set_xlabel('Husband',fontsize=12)
ax.set_ylabel('wife',fontsize=12)
ax.set_title('Scatter Plot of Heights')
a,b = np.polyfit(x,y,deg=1)
y_est = a * x + b
ax.plot(x,y_est,'-')
```

```python
import math
x = np.arange(-5,5,0.1)
y1 = -(1/3)*x+1.5
y2 = 0.02*x*(x-1)*(x+1)+1
y3 = np.array([])
for a in x: y3 = np.append(y3,2*math.exp(-a*a))
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax1.plot(x,y1,color='g',linestyle=':')
ax2.plot(x,y2,color='y',linestyle='--')
ax3.plot(x,y3,color='r',linestyle='-')
```
