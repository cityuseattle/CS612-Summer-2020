```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('toy_dataset1.csv', encoding='latin')
df
```

```python
# 1. Create a new column named Age_group in df by using pd.cut.
#   bins: [0, 30, 45, 60, 75]
#   labels: [ 1, 2, 3, 4]
df['Age_group'] = pd.cut(df.Age, [0,30,45, 60,75], labels=['1','2','3','4'])
df
```

```python
# 2. Create a 2x4 categorical barplot using sns.catplot
#   x-axis: Gender
#   y-axis: Income
#   hue: Illness
#   row: City
#   col=Age_group

sns.catplot(x='Gender', y='Income', hue='Illness', row='City', col='Age_group', data=df)
```
