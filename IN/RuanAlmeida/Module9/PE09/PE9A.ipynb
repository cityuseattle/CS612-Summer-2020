```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. Use pd.read_csv to read the file (toy_dataset1.csv) in the Jupyter Notebook.
df = pd.read_csv('toy_dataset1.csv', encoding='latin')
df
```

```python
# 2. Use sns.scatterplot to draw a scatter plot between Age and Income.
#    Make sure Age appears in the horizontal axis and Income in the vertical axis.
#    Use different colors for gender by using hue.

sns.set(color_codes=True)
sns.scatterplot(x='Age',y='Income', hue='Gender', data=df)
```

```python
# 3. Plot a density function on Income

df['Income'].plot.density() 
```
