```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. Use np.arange() to generate a sequence from -5 to 5 (5 not inclusive) in an increment of 0.1. Call it x
x = np.arange(start=-5, stop=5, step=0.1)
x
```

```python
# 2a.Plotting (Graph 1) y = -(1/3)* x + 1.5
# 3. Make sure the horizontal axis in the plot ranges from -5 to 4.9
y = -(1/3)*x+1.5

# 4a. Use the following line style: green, dotted line
plt.plot(x, y, color='green', linestyle='dotted')

# 5. Put the legend of Graph 1 at the ‘upper center’
plt.legend(["Graph 1"], loc="upper center")
```

```python
# 2b.Plotting (Graph 2) y = 0.02 * x * (x-1) * (x+1) + 1
# 3. Make sure the horizontal axis in the plot ranges from -5 to 4.9
y = 0.02*x*(x-1)*(x+1)+1

# 4b. Use the following line style: yellow, dashed line
plt.plot(x, y, color='yellow', linestyle='dashed')

# 5. Put the legend of Graph 1 at the ‘upper center’
plt.legend(["Graph 2"], loc="upper center")
```

```python
# 2c.Plotting (Graph 3) Y = 2 * exp(-x^2)
# 3. Make sure the horizontal axis in the plot ranges from -5 to 4.9
y = 2*np.exp(-x**2)

# 4c. Use the following line style: red, solid line
plt.plot(x, y, color='red', linestyle='solid')

# 5. Put the legend of Graph 3 at the ‘upper center’
plt.legend(["Graph 3"], loc="upper center")
```
