```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. Use pd.read_csv to read the file (Heights.csv) in the Jupyter Notebook.
Heights = pd.read_csv('Heights.csv', encoding='latin')
Heights
```

```python
# 2. Use matplotlib to draw a scatter plot between husband’s height and wife’s height. (Make sure the husband’s height is in the horizontal axis and the wife’s in the vertical axis
H_Axis = Heights['Husband']
V_Axis = Heights['Wife']

plt.scatter(H_Axis, V_Axis)
plt.show()
```

```python
# 3. Label the horizontal axis “Husband” and the vertical one “Wife”
# 4. Label the title of the plot “Scatter Plot of Heights”
# 5. Draw a fitted line that indicates a linear relationship between these two variables

sns.set(color_codes=True)
ax = sns.regplot(x="Husband", y="Wife", data=Heights, label="Scatter Plot of Heights")
ax.legend(loc="best")
```
