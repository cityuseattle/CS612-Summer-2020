```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
# load tips dataset from seaborn
df = sb.load_dataset('tips')
sb.relplot(x="total_bill", y="tip", data=df)
# scatter is default, add this para kind="line" changes to line plot
```

```python
sb.relplot(x="total_bill", y="tip", hue=df.day.to_list(), data=df)
```

```python
sb.set(style='whitegrid')
sb.relplot(x="total_bill", y="tip", hue=df.day.to_list(), palette="Greens", data=df)
# six default theme: [deep, muted, pastel, bright, dark, and colorblind]
```

```python
sb.palplot(sb.color_palette("muted"))
```

```python
sb.palplot(sb.color_palette("husl", 8))
```

```python
sb.palplot(sb.color_palette("Blues"))
```

```python
sb.catplot(x="day", y="total_bill", data=df)
```

```python
dist = np.random.normal(loc=5, size=100, scale=2)
sb.distplot(dist)
```

```python
# FacetGrid 
fg = sb.FacetGrid(df, col="time")
fg.map(plt.hist, "tip")
```

```python
# PairGrid
# load iris dataset, then plot in different color
iris = sb.load_dataset("iris")
pg = sb.PairGrid(iris, hue="species")
pg.add_legend()
pg.map(plt.scatter)
```

```python
# heatmap 
# load flights dataset and conver to long-form
flights_long = sb.load_dataset("flights")
flights = flights_long.pivot("month", "year", "passengers")

# draw a heatmap with the numeric values in each cell
f, ax = plt.subplots(figsize=(12, 9)) # size of figure
sb.heatmap(flights, annot=True, fmt="d", linewidths=.5, ax=ax)
```
