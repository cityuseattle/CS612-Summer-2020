```python
#this will be a contour plot
def f(x,y): return (1-x/2+x**5+y**3)*np.exp(-x**2-y**2)
n = 256
x = np.linspace(-3,3,n)
y = np.linspace(-3,3,n)
X,Y = np.meshgrid(x,y)

plt.contourf(X,Y, f(X,Y), 8, alpha= 0.75, cmap= 'jet')
C = plt.contour(X,Y, f(X,Y), 8, colors='black', linewidth=0.5)
plt.show()
```

```python
#scatter plt
n = 512
x = np.random.normal(0,1,n)
y = np.random.normal(0,1,n)

plt.scatter(x,y, color='c')
plt.show()
```

```python
#bar plot
n= 12
x = np.arange(n)
y1 = (1-x/float(n)) * np.random.uniform(0.5,1.0,n)
y2 = (1-x/float(n)) * np.random.uniform(0.5,1.0,n)

plt.bar(x, +y1, facecolor='#9999ff', edgecolor=['green', 'red'])
plt.bar(x, -y2, facecolor='#ff9999', edgecolor='white')

for x,y in zip(x,y1):
    plt.text(x+0.4, y+0.05, f"{y:.2f}", ha='center', va='center')

plt.ylim(-1.25,+1.25)
plt.show()
```

```python
#3D plot
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = Axes3D(fig)

x = np.arange(-4,4,0.25)
y = np.arange(-4,4,0.25)

x,y = np.meshgrid(x,y)
r = np.sqrt(x**2 + y**2)
z = np.sin(r)

ax.plot_surface(x,y,z, rstride=1, cstride=1, cmap='winter')

plt.show()
```
